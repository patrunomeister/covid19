```python
pd.options.mode.chained_assignment = None

#connessione a OpenPuglia per ottenere i dati

url= "https://openpuglia.org/api/?q=getdatapccovid-19&reg=puglia&mode=ts"
a = json.dumps(requests.get(url).json())
df = pd.read_json(a, orient='columns')



df.head()
```

```python
trace0 = go.Scatter(
    x = df['data'],
    y = df['ricoverati con sintomi'],
    mode = 'lines',
    name = 'ricoverati'
)

trace1 = go.Scatter(
    x = df['data'],
    y = df['terapia intensiva'],
    mode = 'lines',
    name = 'terapia intensiva'
)

data = [trace0, trace1]

# Plot and embed in ipython notebook!
py.plotly.iplot(data, title='Covid-19 Puglia', filename='ricoverati')
```
